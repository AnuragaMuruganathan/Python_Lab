```python
students = ["Alice", "Bob", "Charlie", "David", "Eve"]
attendance = {}
print("ğŸ“‹ Automatic Attendance System\n")
for student in students:
    status = input(f"Is {student} present? (y/n): ").strip().lower()
    if status == "y":
        attendance[student] = "Present"
    else:
        attendance[student] = "Absent"
print("\nâœ… Attendance Summary:")
for student, status in attendance.items():
    print(f"{student}: {status}")
```

```python
correct_answers = {
    1: "Paris",
    2: "Mercury",
    3: "Einstein",
    4: "Oxygen",
    5: "Pacific"
}
student_answers = {
    1: "Paris",
    2: "Venus",
    3: "Einstein",
    4: "oxygen",
    5: "Atlantic"
}
score = 0
for q_no, correct in correct_answers.items():
    student = student_answers.get(q_no, "").strip().lower()
    if student == correct.lower():
        score += 1
print(f"Score: {score}/{len(correct_answers)}")
```

```python
from difflib import SequenceMatcher
def check_plagiarism(text1, text2):
    similarity = SequenceMatcher(None, text1, text2).ratio()
    return similarity*100
text1 = input("Oiginal text: ")
text2 = input("Comparing text: ")
similarity_percent = check_plagiarism(text1, text2)
print(f"\n Similarity: {similarity_percent:.2f}%")
if similarity_percent > 80:
    print("High chance of plagiarism!")
elif similarity_percent > 50:
    print("Possible plagiarism, needs review.")
else:
    print("Looks original.")
```

```python
store = {
    "T-shirt": 500,
    "Jeans": 1200,
    "Shoes": 2500,
    "Cap": 300,
    "Watch": 1500
}
cart = {}
print("Welcome to the Online AI Shopping System ğŸ›’")
print("Here are the available items:\n")
for item, price in store.items():
    print(f"{item}: â‚¹{price}")
while True:
    choice = input("\nEnter item to add to cart (or type 'done' to checkout): ").strip()
    if choice.lower() == "done":
        break
    if choice in store:
        qty = int(input(f"Enter quantity of {choice}: "))
        if choice in cart:
            cart[choice] += qty
        else:
            cart[choice] = qty
        print(f"{qty} x {choice} added to cart.")
    else:
        print("Item not found. Please choose from the list.")
print("\nğŸ§¾ Final Bill:")
total = 0
for item, qty in cart.items():
    price = store[item] * qty
    total += price
    print(f"{item} x {qty} = â‚¹{price}")

print(f"\nTotal Amount: â‚¹{total}")
print("Thank you for shopping with us! ğŸ˜Š")
```

```python
import random
teachers = {
    "Mr. Sharma": "Math",
    "Ms. Kapoor": "English",
    "Mr. Khan": "Science",
    "Ms. Verma": "History",
    "Mr. Iyer": "Computer"
}
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
periods_per_day = 5
timetable = {}
for day in days:
    timetable[day] = []
    available_teachers = list(teachers.items())
    random.shuffle(available_teachers)
    for i in range(periods_per_day):
        if not available_teachers:
            available_teachers = list(teachers.items())
            random.shuffle(available_teachers)
        teacher, subject = available_teachers.pop()
        timetable[day].append((subject, teacher))
print("ğŸ“… Weekly Timetable for Teachers:\n")
for day in days:
    print(f"{day}:")
    for i, (subject, teacher) in enumerate(timetable[day], 1):
        print(f"  Period {i}: {subject} - {teacher}")
    print()
```
